```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
datos = {
   "Cliente": [1, 2, 3, 4, 5],
   "Edad": [25, 40, 32, 22, 35],
   "Plan": ["Básico", "Premium", "Estándar", "Básico", "Premium"],
   "Consumo_GB": [5, 50, 10, 7, 45],
   "Reclamos": [1, 0, 2, 3, 0],
   "Estado_Cuenta": ["Pagado", "Pagado", "Moroso", "Moroso", "Pagado"]
}

df = pd.DataFrame(datos)

print(df)
```

```python
print("Info:")
print(df.info())

print("\nDescripción:")
print(df.describe())
```

```python
print("Valores faltantes:")
print(df.isnull().sum())
```

```python
sns.histplot(df["Consumo_GB"], bins=5, kde=True)
plt.title("Distribución del Consumo de Datos")
plt.show()
```

```python
sns.countplot(x=df["Estado_Cuenta"])
plt.title("Estado de Cuenta")
plt.show()
```

## Análisis Exploratorio de Datos (EDA)

Se generan gráficos para entender la distribución de los datos.


## Interpretación de Resultados

A partir del análisis realizado, se pueden observar los siguientes puntos:

- Los clientes con mayor consumo de datos tienden a tener planes Premium.
- Los clientes con estado "Moroso" presentan mayor cantidad de reclamos.
- Existe una posible relación entre el número de reclamos y la morosidad.

Estos patrones pueden ayudar a la empresa a tomar decisiones estratégicas, como mejorar la atención a clientes con muchos reclamos.
